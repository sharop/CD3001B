```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.api import VAR
from scipy.stats import pearsonr
```

```python
def parser(s):
    return datetime.strptime(s, '%Y-%m-%d')
```

```python
shakira_df = pd.read_csv('https://raw.githubusercontent.com/sharop/CD3001B/main/VAR/Ejercicio/Shakira.csv', parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
casio_df = pd.read_csv('https://raw.githubusercontent.com/sharop/CD3001B/main/VAR/Ejercicio/Casio.csv', parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
df = pd.merge(shakira_df, casio_df, on=["date"])
```

```python
shak_casio_df = df.asfreq(pd.infer_freq(df.index))
```

```python
plt.figure(figsize=(12,6))
shakira, = plt.plot(shak_casio_df['shakira'])
casio, = plt.plot(shak_casio_df['casio'], color='red')

for year in range(2022, 2023):
    plt.axvline(datetime(year,1,1), linestyle='--', color='k', alpha=0.5)

plt.legend(['Shakira', 'Casio'], fontsize=16)
```
